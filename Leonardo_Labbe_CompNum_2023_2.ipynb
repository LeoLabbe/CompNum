b) Um avião está sendo monitorado por um radar, os dados da posição são obtidos em coordenadas polares $r$ e $\theta$.
Use os dados da tabela abaixo para determinar os vetores velocidade e aceleração $\vec{v}$ e $\vec{a}$, respectivamente, para os tempos tabelados. Use aproximações por diferenças finitas de segunda ordem e compare os resultados.

A velocidade e aceleração em coordenadas polares são dadas pelas seguintes equações:

$\vec{v} = r'\,\vec{e_r} + r \theta'\,\vec{e_\theta},\quad \vec{a} = (r''-r\theta')\,\vec{e_r} + (r \theta''+2 r'\theta')\,\vec{e_\theta}$,

em que $\vec{e_r}$ e $\vec{e_\theta}$ são vetores unitários na direção radial e transversal, resectivamente.

**Observação**: Sempre que for possível dê preferência às fórmulas centradas.

```python
import numpy as np
import math as m

#valores estabelecidos
y = [0, 0.002, 0.006, 0.012, 0.018, 0.024]
v = [0, 0.287, 0.899, 1.915, 3.048, 4.299]
viscosidade = 1.8 * 10**-5

def CalcAdiantado():
    print("no ponto", v[0], ":", format((v[1] - v[0]) / (y[1] - y[0]) * viscosidade, ".50f"))

def CalcCentrado():
    for i in range(1, len(y) - 1):
        derivadaPrimeira = (v[i+1] - v[i-1]) / (y[i+1] - y[i-1])
        print("no ponto", y[i], ":", format(derivadaPrimeira * viscosidade, ".50f"))

def CalcAtrasado():
    print("no ponto", y[5], ":", format((v[5] - v[4]) / (y[5] - y[4]) * viscosidade, ".50f"))


print("A Tensao de Cisalhamento para o calculo adiantado:")
CalcAdiantado()

print("A Tensao de Cisalhamento para o calculo centrado:")
CalcCentrado()

print("A Tensao de Cisalhamento para o calculo atrasado:")
CalcAtrasado()
```
